```python
plt.figure(figsize=(7,5))
plt.title(" Types of Room with higher cancellation as per average daily rate", fontsize= 15, font='Verdana')  
sns.lineplot(data=Hotel_booking, x='reserved_room_type', y='adr', palette='flare', hue= 'is_canceled', marker='o', ls='--')
plt.legend()
plt.savefig('Types of Room with higher cancellation as per average daily rate.png')
plt.show()
```
